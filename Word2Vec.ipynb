```python
from gensim.models import Word2Vec
import string
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sentence="Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational numbers,irrational numbers, geometrical magnitudes, etc., to all be treated as algebraic objects. It gave mathematics a whole new development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itselfin a way which had not happened before."
sentence = sentence.lower()

punctuations = """`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ""" + string.punctuation
data = "".join([item for item in sentence if item not in punctuations])

data1=word_tokenize(data)
model = Word2Vec([data1],min_count=1)
model.save("Word2cev.model")
for i in data1:
    vt = model.wv[i]
    print("Word : {}\nWord Vectors : {}\n\n".format(i,vt))
```
